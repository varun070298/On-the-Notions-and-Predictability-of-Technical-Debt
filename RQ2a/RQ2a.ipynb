```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', None)
```

```python
Qcc_sq_co_cc = pd.read_csv('Qcc_sq_co_cc.csv')
```

```python
TD = pd.read_csv('TD.csv')
TD
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap((pd.concat([Qcc_sq_co_cc, TD], axis=1, keys=['Qcc_sq_co_cc', 'TD']).corr().loc['Qcc_sq_co_cc', 'TD']), vmin=-1, vmax=1, annot=True)
```

```python
Qcc_quant = pd.read_csv('Qcc_quant.csv')
```

```python
plt.figure(figsize=(15, 40))
sns.heatmap((pd.concat([Qcc_quant, TD], axis=1, keys=['Qcc_quant', 'TD']).corr().loc['Qcc_quant', 'TD']), vmin=-1, vmax=1, annot=True)
plt.savefig('CM.png')
```
